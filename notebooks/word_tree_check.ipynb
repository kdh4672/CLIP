```python
import torch
import numpy as np
tree_path = './resources/imagenet21k_miil_tree.pth'
tree = torch.load(tree_path)
class_tree_list = tree['class_tree_list']
class_names = np.array(list(tree['class_description'].values()))
```

```python
total_tree = open("hierarchy_word.txt", "w")
ct = 0
for list_ in class_tree_list: # [5,6,7]
    blank = []
    for num in list_:
        blank.append(class_names[num])
    total_tree.write(str(blank)+'\n')
    ct+=1
total_tree.close()
```

```python
first_tree = open("hierarchy_first_word.txt", "w")
dict_ = {}
for list_ in class_tree_list: # [5,6,7]
    try:
        dict_[class_names[list_[-1]]] += 1
    except:
        dict_[class_names[list_[-1]]] = 1
# first_tree.close()
x = list(dict_.keys())
for item in x:
    first_tree.write(item + '\n')
    
first_tree.close()
```

```python
second_tree = open("hierarchy_second_word.txt", "w")
dict_ = {}
for list_ in class_tree_list: # [5,6,7]
    if len(list_) == 1:
        try:
            dict_[class_names[list_[-1]]] += 1
        except:
            dict_[class_names[list_[-1]]] = 1
    else:
        try:
            dict_[class_names[list_[-1]]] += 1
        except:
            dict_[class_names[list_[-1]]] = 1

x = list(dict_.keys())
for item in x:
    second_tree.write(item + '\n')
    
second_tree.close()
```

```python
full_item = open("full_word.txt", "w")
dict_ = {}
for list_ in class_tree_list: # [5,6,7]
    for item in list_:
        try:
            dict_[class_names[item]] += 1
        except:
            dict_[class_names[item]] = 1

x = list(dict_.keys())
for item in x:
    full_item.write(item + '\n')
    
full_item.close()
```

```python
total_tree = open("dog_word.txt", "w")
ct = 0
for list_ in class_tree_list: # [5,6,7]
    blank = []
    for num in list_:
        blank.append(class_names[num])
    if 'dog' in blank:
        total_tree.write('/'.join(blank)+'\n')
        ct+=1
total_tree.close()
```
